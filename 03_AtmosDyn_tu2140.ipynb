# Question 1.

##(a)

The phase $\phi$ is defined as

$$\phi \equiv kx + ly + mz - \omega z$$

so along constant phase, we get

$$0 = d \phi = \frac{\partial \phi}{\partial x} dx + \frac{\partial \phi}{\partial y} dy + \frac{\partial \phi}{\partial z} dz + \frac{\partial \phi}{\partial t} dt = k \ dx + l \ dy + m \ dz - \omega \ dt$$

$$\therefore 0 = k \frac{dx}{dz} + l \frac{dy}{dz} + m - \omega \frac{dt}{dz}$$

For a steady state on a constant $y$-plane, we have

$$0 = \frac{dx}{dz} + \frac{m}{k}$$

A westward displacement of the phase with increasing height is equivalently $\frac{dx}{dz} < 0$ so from the relation above, we need $\frac{m}{k} > 0 \ \ \ (mk > 0)$.

## (b)

We have

$$v' = \frac{\partial \Psi'}{\partial x}, \ \ \ \ b' = f_0 \frac{\partial \Psi'}{\partial z}$$

so 

$$v' = Re \big[ ik \tilde{\Psi} e^{i\phi} \big] = Re \big[ ik \big(\tilde{\Psi}_r + i\tilde{\Psi}_i\big) (\cos\phi + i\sin\phi) \big]$$ 

$$ = -k \big( \tilde{\Psi}_r \sin\phi + \tilde{\Psi}_i \cos\phi \big)$$

$$b' = -m f_0 \big( \tilde{\Psi}_r \sin\phi + \tilde{\Psi}_i \cos\phi \big)$$

Thus,

$$v'b' = mkf_0 \big( \tilde{\Psi}_r \sin\phi + \tilde{\Psi}_i \cos\phi \big)^2$$ 

$$= mkf_0 \big( \tilde{\Psi}_r^2 \sin^2\phi + \tilde{\Psi}_i^2 \cos^2\phi + 2\tilde{\Psi}_r \tilde{\Psi}_i \cos\phi \sin\phi \big)$$

Now, considering 

$$\int_0^{2\pi/k} \sin^2 (kx+ly+mx-\omega t) \ dx = \int_0^{2\pi/k} \frac{1-\cos 2(kx+ly+mx-\omega t)}{2} \ dx$$ 

$$= \bigg[ \frac{1}{2}x - \frac{1}{4k} \sin 2(kx+ly+mx-\omega t) \bigg]_0^{2\pi/k} = \frac{1}{2}$$

The last equality comes from the periodicity of the $\sin$ function, viz. $\sin \big( 2 (2\pi+ly+mz-\omega t) \big) = \sin \big( 2 (ly+mz-\omega t) \big)$. In the same manner,

$$\int_0^{2\pi/k} \cos^2 (kx+ly+mx-\omega t) \ dx = \frac{1}{2}$$

$$\int_0^{2\pi/k} \sin (kx+ly+mx-\omega t) \cos (kx+ly+mx-\omega t) \ dx = 0$$

so we get,

$$\overline{v'b'} = \int_0^{2\pi/k} v'b' \ dx = \frac{mkf_0}{2} \big( \tilde{\Psi}_r^2+\tilde{\Psi}_i^2 \big) = \frac{mkf_0}{2} \big| \tilde{\Psi} \big|^2$$

Since, westward tilting implies $km > 0$, $\overline{v'b'}$ is positive and from the Eliassen-Palm flux, this is equivalent to a poleward buoyancy flux.

## (c)

The vertical group velocity is

$${\boldsymbol{c}_g}^z \equiv \frac{\partial \omega}{\partial m} = \frac{\partial}{\partial m} \frac{- \beta k}{k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2} = \frac{\beta k \cdot 2\frac{{f_0}^2}{N^2} m}{k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2} = \frac{2\beta k \frac{{f_0}^2}{N^2} }{k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2} \ km$$

Since the necessary condition for poleward buoyancy flux is $km > 0$, we have ${\boldsymbol{c}_g}^z > 0$, an upward group velocity.

## (d)

In order to unpack the wave activity, we need

$$q' = \boldsymbol{\nabla}^2 \Psi' + \frac{{f_0}^2}{N^2} \frac{\partial^2}{\partial z^2} \Psi', \ \ \ \ \bar{q} = \beta y$$

so this yields

$$q' = Re \bigg[ -\bigg( k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2 \bigg) \tilde{\Psi} e^i{\phi} \bigg]$$ 

$$=  -\bigg( k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2 \bigg) \big( \tilde{\Psi}_r \cos\phi - \tilde{\psi}_i \cos\phi \bigg)$$

Thus,

$$\overline{{q'}^2} = \frac{\big( k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2 \big)^2}{2} \big| \tilde{\Psi} \big|^2$$

$$\therefore \mathcal{A} \equiv \frac{1}{2} \frac{\overline{{q'}^2}}{\bar{q}_y} = \frac{\big( k^2 + l^2 + \frac{{f_0}^2}{N^2} m^2 \big)^2}{4 \beta} \big| \tilde{\Psi} \big|^2$$

Hence, 

$${\boldsymbol{c}_g}^z \mathcal{A} = \frac{{f_0}^2}{N^2} \frac{km}{2} \big| \tilde{\Psi} \big|^2 = \frac{f_0}{N^2} \overline{v'b'} = \boldsymbol{F}_z$$

# Question 2.

The streamfunction is defined as

$$\psi' (x, y, t) = e^{i(kx - ly + mz - \omega t)} + R \ e^{i(kx + ly + mz - \omega t)}$$

so we get

$$u' = - \frac{\partial \psi'}{\partial y} = il \big(e^{i(kx - ly + mz - \omega t)} - R \ e^{i(kx + ly + mz - \omega t)} \big)$$

$$v' = \frac{\partial \psi'}{\partial y} = ik \big(e^{i(kx - ly + mz - \omega t)} + R \ e^{i(kx + ly + mz - \omega t)} \big)$$

Taking the real parts of the velocity and then forming a product gives

$$u' v' = - kl \big( - \sin(kx - ly + mz - \omega t) + R \ \sin(kx + ly + mz - \omega t) \big) \big( \sin(kx - ly + mz - \omega t) + R \ \sin(kx + ly + mz - \omega t) \big)$$

$$= kl \big( \sin^2{(kx - ly + mz - \omega t)} - R^2 \sin^2{(kx + ly + mz - \omega t)} \big)$$

Now, if we take the zonal average for each term

$$\frac{k}{2 \pi} \int_0^{\frac{2 \pi}{k}} \sin^2(kx-ly+mz-\omega t) = \frac{k}{2 \pi} \int_0^{\frac{2 \pi}{k}} \frac{1 - \cos \big( 2 (kx-ly+mz-\omega t) \big)}{2}$$

$$= \frac{k}{2 \pi} \bigg[ \frac{1}{2} x - \frac{1}{4k} \sin \big( 2 (kx-ly+mz-\omega t) \big) \bigg]_0^{\frac{2 \pi}{k}} = \frac{1}{2}$$

The last equality comes from the periodicity of the $\sin$ function, viz. $\sin \big( 2 (2\pi-ly+mz-\omega t) \big) = \sin \big( 2 (-ly+mz-\omega t) \big)$. The zonal average of the second term also becomes $1/2$ so we get

$$\overline{u' v'} = \frac{kl}{2} \big( 1 - |R|^2 \big)$$

### Comments

If we separate the meridional flux of momentum into two terms that propagate towards the equator and waves reflected back poleward as $\overline{u' v'}_{eq}$ and $\overline{u' v'}_{pole}$ respectively, we have

$$\overline{u' v'}_{eq} = \frac{kl}{2} \ \ (> 0), \ \ \ \overline{u' v'}_{pole} = - \frac{kl}{2} |R|^2 \ \ (< 0)$$

where the subscripts "eq" and "pole" describe the direction of propagation, so the meridional flux of momentum for propagating waves toward to the critical level is always positive, which acts on to accelerate the jet, and visa versa for the relected waves. When the propagating waves are completely absorbed at the critical level, the net meridional flux of momentum converges to the momentum flux of the propagating waves, viz.

$$\overline{u' v'} = \overline{u' v'}_{eq}$$

so the jet is accelerated by the Rossby wave momentum flux for this case as well.

# Question 3.

Starting off with the buoyancy equation

$$\bigg( \frac{\partial}{\partial t} - \frac{\partial \psi}{\partial y} \frac{\partial}{\partial x} + \frac{\partial \psi}{\partial x} \frac{\partial}{\partial y} \bigg) b + w N^2 = Q_h$$

the advective term can be transformed as

$$\bigg( - \frac{\partial \psi}{\partial y} \frac{\partial}{\partial x} + \frac{\partial \psi}{\partial x} \frac{\partial}{\partial y} \bigg) b = - \frac{\partial}{\partial x} \bigg( \frac{\partial \psi}{\partial y} b \bigg) + \frac{\partial}{\partial y} \bigg( \frac{\partial \psi}{\partial x} b \bigg) + b \bigg( \frac{\partial}{\partial x} \frac{\partial \psi}{\partial y} - \frac{\partial}{\partial y} \frac{\partial \psi}{\partial x} \bigg)$$ $$= - \frac{\partial}{\partial x} \bigg( \frac{\partial \psi}{\partial y} b \bigg) + \frac{\partial}{\partial y} \bigg( \frac{\partial \psi}{\partial x} b \bigg)$$

Thus the buoyancy equation becomes

$$\frac{\partial}{\partial t} b - \frac{\partial}{\partial x} \bigg( \frac{\partial \psi}{\partial y} b \bigg) + \frac{\partial}{\partial y} \bigg( \frac{\partial \psi}{\partial x} b \bigg) + w N^2 = Q_h \ \ \ \ \ \ (3-1)$$

and separating each variable into the time mean and perturbation, taking the zonal average gives

$$\overline{\frac{\partial}{\partial t} (\bar{b} + b') - \frac{\partial}{\partial x} \bigg[ \bigg(\frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg] + \frac{\partial}{\partial y} \bigg[ \bigg(\frac{\partial \bar{\psi}}{\partial x} + \frac{\partial \psi'}{\partial x} \bigg) (\bar{b} + b') \bigg] + (\bar{w} + w') N^2 = \overline{Q}_h+Q_h'}$$

By periodicity, $\overline{\frac{\partial}{\partial x}(\ )} = 0$, so the second term and $\frac{\partial \bar{\psi}}{\partial x}$ on the left-hand side in the equation above drops out. Also using the fact that time mean of the perturbation terms are zero gives

$$\frac{\partial}{\partial t} \bar{b} + \frac{\partial}{\partial y} \bigg(\overline{\frac{\partial \psi'}{\partial x} b'} \bigg) + \bar{w} N^2 = \bar{Q}_h \ \ \ \ \ \ (3-2)$$

Multiplying both sides with $\frac{\tilde{\rho}}{N^2} \bar{b}$ yields

$$\frac{\partial}{\partial t} \big( \frac{1}{2} \frac{\tilde{\rho}}{N^2} \bar{b}^2 \big) = - \frac{\tilde{\rho}}{N^2} \bar{b}\frac{\partial}{\partial y} \bigg(\overline{\frac{\partial \psi'}{\partial x} b'} \bigg) - \tilde{\rho} \bar{w} \bar{b} + \frac{\tilde{\rho}}{N^2} \bar{Q}_h \bar{b}$$

Taking the integral over $y$ and $z$ gives

$$\frac{\partial \bar{P}}{\partial t} = \iint \frac{\tilde{\rho}}{N^2} \bar{b}\frac{\partial}{\partial y} \bigg(- \overline{v' b'} \bigg) \ dy dz - \iint \tilde{\rho} \bar{w} \bar{b} \ dy dz + \iint \frac{\tilde{\rho}}{N^2} \bar{Q}_h \bar{b} \ dy dz$$

which is the equation for the $\textbf{potential energy of the zonal mean state}$.

Now, subtracting equation (3-2) from (3-1) gives

$$\frac{\partial}{\partial t} b' - \frac{\partial}{\partial x} \bigg[ \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg] + \frac{\partial}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg] = \frac{\partial}{\partial y} \bigg(\overline{\frac{\partial \psi'}{\partial x} b'} \bigg) - w' N^2 + Q_h'$$

and we shall multiply both sides by $b'$ and take the zonal average which gives

$$\overline{ \frac{\partial}{\partial t} \bigg(\frac{1}{2} b'^2 \bigg) - b' \frac{\partial}{\partial x} \bigg[ \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg] + b' \frac{\partial}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg] = b' \frac{\partial}{\partial y} \bigg(\overline{v' b'} \bigg) - w' b' N^2 + Q_h' b' } \ \ \ \ \ (3-3)$$

First consider

$$- \overline{b' \frac{\partial}{\partial x} \bigg[ \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]} = - \overline{\frac{\partial}{\partial x} \bigg[b' \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]} + \overline{\bigg[\frac{\partial b'}{\partial x} \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]}$$

$$= \overline{\bigg[\frac{\partial b'}{\partial x} \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]}$$

$$= \overline{\frac{\partial b'}{\partial x} \frac{\partial \bar{\psi}}{\partial y} (\bar{b} + b')} + \overline{\frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y} (\bar{b} + b') }$$

$$= \bar{b} \frac{\partial \bar{\psi}}{\partial y}\frac{\partial \overline{b'}}{\partial x} + \frac{\partial \bar{\psi}}{\partial y} \overline{ \frac{\partial}{\partial x} \bigg( \frac{1}{2} b'^2 \bigg)} + \bar{b} \overline{\frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y}} + \overline{ b' \frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y} }$$

$$= \bar{b} \overline{\frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y}} + \overline{ b' \frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y} }$$

where we have used $\overline{\frac{\partial}{\partial x} (\ )} = 0$ and $\overline{(\ ')} = 0$.

Now consider

$$\overline{ b' \frac{\partial}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg] } = \frac{\partial}{\partial y} \overline{ b' \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg]} - \overline{\frac{\partial b'}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg]} $$

$$= \frac{\partial}{\partial y} \overline{ b' \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg]} - \bar{b} \overline{\frac{\partial b'}{\partial y} \frac{\partial \psi'}{\partial x}} - \overline{ b' \frac{\partial b'}{\partial y} \frac{\partial \psi'}{\partial x}}$$

Integrating the two terms above over $y$ gives

$$\int - \overline{b' \frac{\partial}{\partial x} \bigg[ \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]} \ dy = \int \bar{b} \overline{\frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y}} \ dy + \int \overline{ b' \frac{\partial b'}{\partial x} \frac{\partial \psi'}{\partial y} }\ dy$$

$$= \int \bigg[ -\bar{b} \overline{\frac{\partial}{\partial x} (u' b')} + \bar{b} \ \overline{b' \frac{\partial u'}{\partial x}} \bigg] \ dy + \int \bigg[ -\overline{\frac{\partial}{\partial x} \big( \frac{1}{2} u' {b'}^2 \big)} + \frac{1}{2} \overline{{b'}^2 \frac{\partial u'}{\partial x}} \bigg] \ dy$$

$$= \int \bar{b} \ \overline{b' \frac{\partial u'}{\partial x}} \ dy + \int \frac{1}{2} \overline{{b'}^2 \frac{\partial u'}{\partial x}} \ dy \ \ \ \ \ (3-4)$$

and

$$\int \overline{ b' \frac{\partial}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg] } \ dy = - \int \bar{b} \overline{\frac{\partial b'}{\partial y} \frac{\partial \psi'}{\partial x}} \ dy - \int \overline{ b' \frac{\partial b'}{\partial y} \frac{\partial \psi'}{\partial x}} \ dy$$

$$= \int \bigg[ - \bar{b} \ \frac{\partial}{\partial y} \big( \overline{v'b'} \big) + \bar{b} \ \overline{b'\frac{\partial v'}{\partial y}} \bigg] \ dy + \int \bigg[ - \frac{\partial}{\partial y} \big( \frac{1}{2} \overline{v'{b'}^2} \big) + \frac{1}{2} \overline{{b'}^2 \frac{\partial v'}{\partial y}} \bigg] \ dy$$

$$= - \int \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy + \int \bar{b} \ \overline{b'\frac{\partial v'}{\partial y}} \ dy + \int \frac{1}{2} \overline{{b'}^2 \frac{\partial v'}{\partial y}} \ dy \ \ \ \ \ (3-5)$$

since $v'$ is zero at $y=0,L$. Thus, summing equations (3-4) and (3-5) gives

$$\int - \overline{b' \frac{\partial}{\partial x} \bigg[ \bigg( \frac{\partial \bar{\psi}}{\partial y} + \frac{\partial \psi'}{\partial y} \bigg) (\bar{b} + b') \bigg]} \ dy + \int \overline{ b' \frac{\partial}{\partial y} \bigg[ \frac{\partial \psi'}{\partial x} (\bar{b} + b') \bigg] } \ dy$$ 

$$= \int \bar{b} \ \overline{b' \frac{\partial u'}{\partial x}} \ dy + \int \frac{1}{2} \overline{{b'}^2 \frac{\partial u'}{\partial x}} \ dy - \int \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy + \int \bar{b} \ \overline{b'\frac{\partial v'}{\partial y}} \ dy + \int \frac{1}{2} \overline{{b'}^2 \frac{\partial v'}{\partial y}} \ dy $$

$$= -\int \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy + \int \bar{b} \ \overline{b' \big( \frac{\partial u'}{\partial x} + \frac{\partial v'}{\partial y} \big)} \ dy + \int \frac{1}{2} \overline{{b'}^2 \big( \frac{\partial u'}{\partial x} + \frac{\partial v'}{\partial y} \big)} \ dy$$

$$= -\int \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy \ \ \ \ \ (3-5)$$

due to mass conservation, viz. $\boldsymbol{\nabla}_h \cdot \boldsymbol{u} = 0$.

Now, the first term on the right-hand side of equation (3-3) is zero because

$$\overline{b' \frac{\partial}{\partial y} \big(\overline{v' b'} \big)} = \overline{b'} \frac{\partial}{\partial y} \big(\overline{v' b'} \big) = 0$$

so using the equation above and (3-5), equation (3-3) ends up to be

$$\int \frac{\partial}{\partial t} \bigg(\frac{1}{2} \overline{b'^2} \bigg) \ dy = \int \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy - \int \overline{w' b'} N^2 \ dy + \int \overline{Q_h' b'} \ dy \ \ \ \ \ (3-6)$$

Multiplying both sides of (3-6) with $\tilde{\rho} / N^2$ and taking an integral over $z$ gives

$$\frac{\partial P'}{\partial t} = \iint \frac{\tilde{\rho}}{N^2} \bar{b} \frac{\partial}{\partial y} (\overline{v' b'}) \ dy \ dz - \iint \tilde{\rho} \overline{w' b'} \ dy \ dz + \iint \frac{\tilde{\rho}}{N^2} \overline{Q_h' b'} \ dy \ dz$$

which is the equation for $\textbf{eddy potential energy}$.
